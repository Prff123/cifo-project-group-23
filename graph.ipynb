```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math 
```

```python
df_100gens_200pop = pd.read_excel("Results\Original\df_100gens_200pop.xlsx")
df_100gens_500pop = pd.read_excel("Results\Original\df_100gens_500pop.xlsx")
df_300gens_200pop = pd.read_excel("Results\Original\df_300gens_200pop.xlsx")
df_300gens_500pop = pd.read_excel("Results\Original\df_300gens_500pop.xlsx")
```

```python
# Extract the column names corresponding to average best fitness
fitness_columns = df_100gens_200pop.columns[1::3]
# Plot the lines and fill the confidence interval for each fitness column
colors=["red","lime","black","grey","darkviolet","magenta","turquoise","chocolate","blue"]
for col,color in zip(fitness_columns,colors):
    avg_fitness = df_100gens_200pop[col]
    std_fitness = df_100gens_200pop[col.replace('avg_best_fitness', '_std_best_fitness')]
    upper_bound = avg_fitness + 1.96 * (std_fitness/math.sqrt(75))
    lower_bound = avg_fitness - 1.96 * (std_fitness/math.sqrt(75))
    
    # Extract the parameter settings from the column name
    parameters = col.split(', ')[1:-1]
    label = ', '.join(parameters)
    
    # Plot the line
    plt.plot(df_100gens_200pop['Gens'], avg_fitness, label=label, color=color)
    
    # Fill the confidence interval area
    plt.fill_between(df_100gens_200pop['Gens'], lower_bound, upper_bound, alpha=0.2, color=color)

# Set labels and title
plt.xlabel('Generation')
plt.ylabel('Average Best Fitness')
plt.title('nr_generations:100 nr_population:200')

# Move the legend outside the graph and adjust its position
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show grid
plt.grid(False)

# Adjust the layout to prevent overlapping of legend and graph
plt.tight_layout()

# Display the graph
plt.show()
```

```python
best_4 = pd.read_excel("Results/Original/best_4.xlsx")
best_4.head()
```

```python
plt.figure(figsize=(8, 7))
# Extract the column names corresponding to average best fitness
fitness_columns = best_4.columns[1::3]

runs = [75,30,75,30]
r = 0
gens = [100,100,300,300]

# Plot the lines and fill the confidence interval for each fitness column
for col, gen in zip(fitness_columns,gens):
    avg_fitness = best_4[col]
    std_fitness = best_4[col.replace('avg_best_fitness', '_std_best_fitness')]
    
    upper_bound = avg_fitness + 1.96 * (std_fitness/math.sqrt(runs[r]))
    lower_bound = avg_fitness - 1.96 * (std_fitness/math.sqrt(runs[r]))
    r+=1
    
    # Extract the parameter settings from the column name
    parameters = col.split(', ')
    label = f"{parameters[0]}, nr_gens={gen}"  # Only include first and third parameters
    
    # Plot the line
    plt.plot(best_4['Gens'], avg_fitness, label=label)
    
    # Fill the confidence interval area
    plt.fill_between(best_4['Gens'], lower_bound, upper_bound, alpha=0.2)

# Set labels and title
plt.xlabel('Generation', fontsize=14)
plt.ylabel('Average Best Fitness', fontsize=14)
plt.title('Comparison of the best models by nr_generations and population_size', fontsize=14)

# Move the legend outside the graph and adjust its position
plt.legend(loc='lower right', fontsize=12)


# Show grid
plt.grid(False)

# Adjust the layout to prevent overlapping of legend and graph
plt.tight_layout()

# Display the graph
plt.show()
```

```python
pop_exploration = pd.read_excel("Results\Original\pop_exploration_300.xlsx")
pop_exploration.head()
```

```python
plt.figure(figsize=(12, 8))

# Extract the column names corresponding to average best fitness
fitness_columns = pop_exploration.columns[1::3]

runs = [75,50,50,30,75,50,50,30]
r = 0

# Plot the lines and fill the confidence interval for each fitness column
for col in fitness_columns:
    avg_fitness = pop_exploration[col]
    std_fitness = pop_exploration[col.replace('avg_best_fitness', '_std_best_fitness')]
    
    upper_bound = avg_fitness + 1.96 * (std_fitness/math.sqrt(runs[r]))
    lower_bound = avg_fitness - 1.96 * (std_fitness/math.sqrt(runs[r]))
    r+=1
    
    # Extract the parameter settings from the column name
    parameters = col.split(', ')[0:-2]
    label = ', '.join(parameters)
    
    # Plot the line
    plt.plot(pop_exploration['Gens'], avg_fitness, label=label)
    
    # Fill the confidence interval area
    plt.fill_between(pop_exploration['Gens'], lower_bound, upper_bound, alpha=0.2)

# Set labels and title
plt.xlabel('Generation', fontsize=15)
plt.ylabel('Average Best Fitness', fontsize= 15)
plt.title('Comparison of the best models by Population Size', fontsize=16)

# Draw a horizontal line for the maximum y value
max_y = pop_exploration[fitness_columns].max().max()
plt.axhline(max_y, color='b', linestyle='--', label='Max Average Best Fitness')

# Add a label for the maximum y-value on the y-axis
plt.annotate(f'Max: {max_y:.0f}', xy=(0, max_y), xytext=(-35, 0),
             textcoords='offset points', ha='right', va='center', color='b', fontsize=12)


# Move the legend outside the graph and adjust its position
plt.legend(loc='lower right', fontsize=14)

# Show grid
plt.grid(False)

# Adjust the layout to prevent overlapping of legend and graph
plt.tight_layout()

# Display the graph
plt.show()
```
